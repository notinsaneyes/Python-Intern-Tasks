```python
import pandas as pd

# Load the dataset with a different encoding
df = pd.read_csv('sales_data.csv', encoding='ISO-8859-1')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Filtering data: Get sales records for Category 1
category_1_sales = df[df['Branch'] == 'branch1']
print(category_1_sales)
```

```python
# Sorting data: Sort by Quantity in descending order
sorted_df = df.sort_values(by='Quantity', ascending=False)
print(sorted_df)
```

```python
# Grouping data: Group by Category and calculate the total Quantity
grouped_df = df.groupby('Branch')['Quantity'].sum().reset_index()
print(grouped_df)
```

```python

# Calculate summary statistics for the Quantity column
mean_quantity = df['Quantity'].mean()
median_quantity = df['Quantity'].median()
std_quantity = df['Quantity'].std()

print(f"Mean Quantity: {mean_quantity}")
print(f"Median Quantity: {median_quantity}")
print(f"Standard Deviation of Quantity: {std_quantity}")
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Visualization: Histogram of Quantity
plt.figure(figsize=(10, 6))
sns.histplot(df['Quantity'], bins=10, kde=True)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization: Bar plot of total Quantity by Branch
plt.figure(figsize=(10, 6))
sns.barplot(x='Branch', y='Quantity', data=grouped_df)
plt.title('Total Quantity by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Quantity')
plt.show()
```

```python

# Visualization: Scatter plot of Quantity vs. Unit Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Unit price', data=df, hue='Branch')
plt.title('Quantity vs. Price by Category')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.legend(title='Category')
plt.show()
```
